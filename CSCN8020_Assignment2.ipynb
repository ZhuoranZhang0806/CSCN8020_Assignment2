## Q-Learning Overview

Q-Learning is a model-free reinforcement learning algorithm. It does not need to know the full transition rules of the environment before training. Instead, the agent learns by interacting with the environment many times.

In the Taxi environment, the agent observes the current state, chooses an action, receives a reward, and moves to a new state. Q-Learning stores learned values in a Q-table. Each value represents how useful it is to take a specific action in a specific state.

During training, the agent updates the Q-table after each action. Over time, better actions receive higher Q-values, so the agent can learn a policy that completes the taxi task more efficiently.

### Q-Learning Update Equation

$$
Q(s, a) = Q(s, a) + \alpha \left[ r + \gamma \max_{a'} Q(s', a') - Q(s, a) \right]
$$

Where:

- $Q(s, a)$ is the current Q-value for taking action $a$ in state $s$.
- $s$ is the current state of the agent.
- $a$ is the action selected by the agent.
- $r$ is the reward received after taking the action.
- $s'$ is the next state after the action is taken.
- $a'$ represents possible actions in the next state.
- $\max_{a'} Q(s', a')$ is the best estimated future Q-value from the next state.
- $\alpha$ is the learning rate. It controls how much new information changes the old Q-value.
- $\gamma$ is the discount factor. It controls how much the agent cares about future rewards.

This update helps the agent improve its Q-values over time. If an action leads to better future rewards, its Q-value will increase. If an action leads to poor results, its Q-value will become lower.

### States

A state represents the current situation of the environment. In the Taxi environment, a state includes the taxi position, passenger location, and destination.

The total number of states is:

$$

$$

The taxi position is based on a 5 by 5 grid. The row number goes from top to bottom, and the column number goes from left to right. A taxi position is written as `(row, column)`.

The four main locations are:

| Color  | Number | Grid Position |

| Green  |      1 | `(0, 4)`      |

The passenger locations are:

The destination locations are:

| Number | Meaning |

$$

### Action Space

The Taxi environment has 6 possible actions:

| Action Number | Action Meaning                               |
| ------------: | -------------------------------------------- |
|             0 | Move south, which means move down one row    |
|             1 | Move north, which means move up one row      |
|             2 | Move east, which means move right one column |
|             3 | Move west, which means move left one column  |
|             4 | Pick up passenger                            |
|             5 | Drop off passenger                           |

The agent chooses one action at each step. Some movements may be blocked by walls in the Taxi map, so the taxi may not always move even if a movement action is selected.

### Rewards

A reward is the feedback from the environment after the agent takes an action. The reward tells the agent whether the action was good or bad.

The reward structure is:

| Situation                                      | Reward |
| ---------------------------------------------- | -----: |
| Successful drop-off at the correct destination |    +20 |
| Illegal pickup or illegal drop-off             |    -10 |
| Each normal step                               |     -1 |

The agent should learn to finish the task with fewer steps and avoid illegal pickup or drop-off actions.

### Episodes

An episode is one complete run of the task. In the Taxi environment, an episode starts when the environment is reset. It ends when the passenger is successfully dropped off or when the maximum number of steps is reached.

During each episode, the agent repeatedly observes the current state, chooses an action, receives a reward, and moves to the next state.

### Exploration

Exploration means the agent tries actions that may not currently look like the best action. This is important because the agent needs to discover better actions by trying different choices during training.

For example, at the beginning of training, the Q-values are not useful yet, so the agent needs exploration to learn about the environment.

### Exploitation

Exploitation means the agent chooses the action that currently has the highest Q-value. This allows the agent to use what it has already learned.

In Q-Learning, exploration and exploitation need to be balanced. If the agent only explores, it may act randomly too often. If the agent only exploits, it may miss better actions.

### Q-values

A Q-value estimates how good it is to take a specific action in a specific state. The Q-table stores Q-values for all state-action pairs.

For example, $Q(s, a)$ represents the estimated value of taking action $a$ in state $s$.

At the beginning of training, the Q-values usually start at zero. As the agent interacts with the environment, these values are updated based on rewards and future expected rewards.

### Temporal-Difference Updates

Q-Learning uses temporal-difference updates. This means the agent updates the current Q-value by comparing the old estimate with a new target.

The new target is based on the immediate reward and the best estimated future Q-value from the next state. This allows the agent to learn step by step without needing to know the full environment model in advance.

### Learning Rate

The learning rate is represented by $\alpha$. It controls how much new information changes the old Q-value.

A higher learning rate makes the agent update Q-values more strongly. A lower learning rate makes the agent update Q-values more slowly.

### Discount Factor

The discount factor is represented by $\gamma$. It controls how much the agent cares about future rewards.

A higher discount factor means the agent gives more importance to future rewards. A lower discount factor means the agent focuses more on immediate rewards.

### Policy Improvement

A policy is the agent's strategy for choosing actions. In Q-Learning, the policy improves as the Q-values are updated.

If an action leads to better rewards, its Q-value becomes higher. Then the agent becomes more likely to choose that action in the future. Through many episodes, the agent gradually learns a better policy for completing the Taxi task.

#### Combined Training Comparison Plots Observations

### Observations from Combined Training Comparison Plots

From the combined training comparison plots, the learning rate has a strong effect on how fast the agent learns. The baseline and `alpha = 0.2` curves improve much faster than the smaller learning rate experiments. In the smoothed return plot, both baseline and `alpha = 0.2` move close to 0 quickly, which means the agent receives fewer negative rewards and learns a better policy.

The `alpha = 0.2` experiment shows the fastest early improvement. Its return increases quickly, its steps decrease quickly, and its average reward per step becomes positive early. This means a higher learning rate allows the Q-table to update faster. In this experiment, `alpha = 0.2` is still stable and helps the agent learn efficiently.

The `alpha = 0.01` experiment improves, but much more slowly than the baseline. Its smoothed return is still lower, and its steps remain higher for many episodes. This means the learning rate is too small, so the Q-values are updated slowly.

The `alpha = 0.001` experiment performs the worst. Its return stays very low, its steps stay very high, and its average reward per step stays negative. This shows that the learning rate is too small for the agent to learn a useful policy within 5000 episodes.

The epsilon experiments show that increasing exploration still allows the agent to learn, but the curves become less efficient than the baseline. `epsilon = 0.2` and `epsilon = 0.3` both improve over time, but their average reward per step is lower than baseline and `alpha = 0.2`. This happens because higher epsilon makes the agent take more random actions during training. Random actions help exploration, but they also cause more inefficient moves and more penalties.

Overall, the combined training plots show that `alpha` mainly affects learning speed, while `epsilon` mainly affects exploration and training noise. A very small alpha slows learning too much. A higher epsilon increases exploration, but too much randomness can reduce training efficiency. The best training behavior comes from the baseline and `alpha = 0.2`.
