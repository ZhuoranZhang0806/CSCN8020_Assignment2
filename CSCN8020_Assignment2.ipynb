## Q-Learning Overview

Q-Learning is a model-free reinforcement learning algorithm. It does not need to know the full transition rules of the environment before training. Instead, the agent learns by interacting with the environment many times.

In the Taxi environment, the agent observes the current state, chooses an action, receives a reward, and moves to a new state. Q-Learning stores learned values in a Q-table. Each value represents how useful it is to take a specific action in a specific state.

During training, the agent updates the Q-table after each action. Over time, better actions receive higher Q-values, so the agent can learn a policy that completes the taxi task more efficiently.

### Q-Learning Update Equation

$$
Q(s, a) = Q(s, a) + \alpha \left[ r + \gamma \max_{a'} Q(s', a') - Q(s, a) \right]
$$

Where:

- $Q(s, a)$ is the current Q-value for taking action $a$ in state $s$.
- $s$ is the current state of the agent.
- $a$ is the action selected by the agent.
- $r$ is the reward received after taking the action.
- $s'$ is the next state after the action is taken.
- $a'$ represents possible actions in the next state.
- $\max_{a'} Q(s', a')$ is the best estimated future Q-value from the next state.
- $\alpha$ is the learning rate. It controls how much new information changes the old Q-value.
- $\gamma$ is the discount factor. It controls how much the agent cares about future rewards.

This update helps the agent improve its Q-values over time. If an action leads to better future rewards, its Q-value will increase. If an action leads to poor results, its Q-value will become lower.

## Taxi Environment and Code Structure

In this project, the Taxi environment is created by the `TaxiEnvironmentManager` class. This class is responsible for creating the Gymnasium Taxi environment and providing basic information about the environment.

The environment is created in the `create_env()` method. In this method, `gym.make(self.env_name)` is used to create the Taxi environment. In the main program, I create a `TaxiEnvironmentManager` object and then call `create_env()` to initialize the environment.

The action space is also explained in the environment manager. The Taxi environment has 6 possible actions:

| Action Number | Action Meaning     |
| ------------- | ------------------ |
| 0             | Move south         |
| 1             | Move north         |
| 2             | Move east          |
| 3             | Move west          |
| 4             | Pick up passenger  |
| 5             | Drop off passenger |

The observation space is checked through the environment information. The Taxi environment has 500 possible states. Each state represents the taxi row, taxi column, passenger location, and destination. The `decode_observation()` method is used to convert a state number into readable information.

The agent is implemented using an object-oriented structure. The base class `TDAgent` stores common information such as the number of states, number of actions, learning rate, discount factor, exploration rate, and Q-table. The Q-table is initialized with zeros for all state-action pairs.

The `QLAgent` class is used for this assignment. It inherits from `TDAgent` and implements the Q-Learning update rule in the `update_q_values()` method. During training, the agent selects an action, receives a reward, observes the next state, and updates the Q-value.
